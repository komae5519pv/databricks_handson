## 1. マルチエージェントスーパーバイザーを作る

- Agent Bricks -> マルチエージェントスーパーバイザー -> ビルド
- 名前: `ma-inventry-analysis`
- 説明:`店舗・倉庫・オンラインの在庫や入荷予定データを統合し、自然言語で在庫状況を確認できるエージェントです。ユーザーの質問に対応し、在庫・入荷・取り寄せを横断的に分析して最適な回答を提示します。`
- エージェントを設定

| タイプ | エージェントエンドポイント/Genieスペース | エージェント名 | コンテンツを説明 ※人が理解しやすい説明 |
| ------------- | ---------------------------------------- | ---------------------------------------------- | ------------------------------------------------------------------------ |
| Genieスペース | アパレル在庫分析 | agent | inventory / product / reservation テーブルを参照し、在庫数・入荷予定・予約状況などのStructuredデータを横断的に分析するエージェント |
| ナレッジアシスタント | kw-store-knowledge-bot | agent | 店舗運営マニュアル、接客研修資料、CS対応FAQ、ブランドブックを参照し、原則・例外・判断基準・接客トークを整理するUnstructuredナレッジ参照エージェント |

- オプション
  - 手順設定（オプション）: 
    ```
    * アパレルショップの店舗スタッフが使う意思決定支援アシスタントです
    * 必ず日本語で回答してください
    * 最初に結論を簡潔に述べ、その後に詳細を説明してください

    * 在庫数・数量・入荷日などのStructuredデータ確認が必要な場合は「アパレル在庫分析（Genie）」を使用してください
    * ルール・例外・判断基準・接客トークが必要な場合は「kw-store-knowledge-bot（RAG）」を使用してください
    * 両方が必要な場合は、必ず両エージェントを使用してください

    * 最終回答は以下の順で整理してください

      1. 事実（在庫・数量など）
      2. 原則
      3. 例外条件
      4. 推奨アクション
      5. 接客トーク例（必要な場合）

    * 取り置き予約可能か？と聞かれた場合は、
      - Genieで購入可能在庫数と予約数を確認し、
      - RAGで安全在庫基準・例外条件を参照して判断してください

    * 店長判断・本部判断が必要な場合は明示してください
    ```

## 2. マルチエージェントスーパーバイザーを使ってみる

|質問番号|質問内容|主な参照元（エージェント）|狙い・意図|
|----|----------------------------------------------|------------------------------|----------------------------------------------|
|Q1|都内で商品ID 1008 ある？在庫がなければ最短入荷も教えて。色はネイビーです|Genie主導|Structuredデータのみで完結する質問。在庫・店舗横断・入荷予定を正確に返せるか確認する。|
|Q2|商品ID 1008のネイビーは、新宿店で取り置き予約できる？|Genie + RAG|Genieで購入可能在庫と予約状況を確認し、RAGで安全在庫基準・例外条件を参照する。数値＋ルール統合の確認。|
|Q3|その商品について教えて|RAG主導（Genie補助）|商品マスタ情報に加え、ブランドブック・接客研修資料を参照してトーンを整えた説明ができるか確認する。|
|Q4|都内の店舗で、商品ID 1008、新宿以外に在庫ある？|Genie主導|店舗横断在庫分析の確認。純粋なStructured検索。|
|Q5|このブランドでレビュー高い商品、在庫あるやつ教えて。|Genie + RAG|Genieでレビュー×在庫抽出、RAGで「なぜ評価が高いか」の訴求ポイント整理。ブランドトーンでの提案ができるか確認。|
