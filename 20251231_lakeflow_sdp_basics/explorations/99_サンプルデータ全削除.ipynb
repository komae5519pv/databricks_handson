## 生成済みのサンプルデータを全て削除する
- **最初からやり直したい場合に、このノートブックを実行してデータを削除してください。**
- スキーマは削除されません。
- スキーマ配下のリソース全て（Declarative Pipelinesで作成したStreaming TableやMetarialized View含む）削除されます。

```python
%run ./00_config
```

```python
# 直下のエントリを列挙
entries = dbutils.fs.ls(volume_dir)

# すべて再帰削除
for e in entries:
    # print(e.path)
    dbutils.fs.rm(e.path, True)
```

```python
# Volume削除
spark.sql(f"DROP VOLUME IF EXISTS {MY_CATALOG}.{MY_SCHEMA}.{MY_VOLUME}")
```

```python
# スキーマ内のすべてのテーブル名を取得する
tables_df = spark.sql(f"SHOW TABLES IN {MY_CATALOG}.{MY_SCHEMA}")

# テーブル名が "silver_" で始まるテーブルのみ削除する
for table in tables_df.collect():
    table_name = table["tableName"]
    spark.sql(f"DROP TABLE IF EXISTS {MY_CATALOG}.{MY_SCHEMA}.{table_name}")
    print(f"削除されたテーブル: {table_name}")

print("全てのテーブルが削除されました。")
```
